```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of sensors
num_of_sensors = 50

# Generate random sensor node positions within a 10x10 km area
np.random.seed(5)
data = 10 + (20 - 10) * np.random.random((num_of_sensors, 2))

# Display the sensor nodes
plt.figure(figsize=(8, 6))
plt.plot(data[:, 0], data[:, 1], 'r.', markersize=10)

# Annotate the sensor nodes with their indices
for i in range(num_of_sensors):
    dx = data[i, 0]
    dy = data[i, 1]
    plt.text(dx - 0.3, dy + 0.3, str(i + 1), fontsize=12)

plt.title('Sensor Nodes Distribution')
plt.xlabel('X-coordinate (km)')
plt.ylabel('Y-coordinate (km)')
plt.grid(True)
plt.show()

# Calculate Euclidean distance between two specific nodes (example)
node1_idx = 0
node2_idx = 1

dist = np.sqrt((data[node2_idx, 0] - data[node1_idx, 0])**2 + (data[node2_idx, 1] - data[node1_idx, 1])**2)
print(f"Euclidean distance between node {node1_idx + 1} and node {node2_idx + 1}: {dist:.2f}")

# Alternatively, compute distances between all pairs of nodes and store in a matrix
dist_matrix = np.zeros((num_of_sensors, num_of_sensors))

for i in range(num_of_sensors):
    for j in range(num_of_sensors):
        dist_matrix[i, j] = np.sqrt((data[j, 0] - data[i, 0])**2 + (data[j, 1] - data[i, 1])**2)

# Print the distance matrix (for illustration, can be large)
print("Distance Matrix:")
print(dist_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the 400 nodes datapoints
num = 400
np.random.seed(42)  # for reproducibility
data = 0 + (100 - 0) * np.random.random((num, 2))

plt.figure(1, (20, 8))

# Raw data plot
plt.subplot(1, 2, 1)
plt.title('Raw data')
plt.plot(data[:, 0], data[:, 1], 'rs')
plt.xlim(0, 100)
plt.ylim(0, 100)

# Clustered data plot
plt.subplot(1, 2, 2)
plt.title("Clustered Data (Four Quadrants)")

for i in range(data.shape[0]):
    x, y = data[i, 0], data[i, 1]
    if x <= 50 and y > 50:
        plt.plot(x, y, 'ro')  # Red for top-left quadrant
    elif x > 50 and y > 50:
        plt.plot(x, y, 'bo')  # Blue for top-right quadrant
    elif x <= 50 and y <= 50:
        plt.plot(x, y, 'go')  # Green for bottom-left quadrant
    else:
        plt.plot(x, y, 'yo')  # Yellow for bottom-right quadrant

# Add quadrant dividing lines

plt.xlim(0, 100)
plt.ylim(0, 100)

# Add legend
plt.plot([], [], 'ro', label='Top-Left')
plt.plot([], [], 'bo', label='Top-Right')
plt.plot([], [], 'go', label='Bottom-Left')
plt.plot([], [], 'yo', label='Bottom-Right')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# auto clustering (2 cluster algorithm )
num=400
data = 0+(100-0)*np.random.random((num,2))
plt.figure(1,(10,2))
plt.subplot(1,2,1)
#plt.title('Raw data')
plt.plot(data[:,0],data[:,1],'rs')

# consider datapoint0 as centroid 1
c1x = data[0,0];c1y = data[0,1]
# consider datapoint1 as centroid 2
c2x = data[1,0];c2y = data[1,1]

# step 2 find the distances of centroid and 1 and centroid 2 with respect to
# all the datapoints

Cluster1x=[];Cluster1y=[]
Cluster2x=[];Cluster2y=[]
for i in range(data.shape[0]):
  d1 = np.sqrt((c1x-data[i,0])**2 + (c1y-data[i,1])**2)
  d2 = np.sqrt((c2x-data[i,0])**2 + (c2y-data[i,1])**2)
  if(d1<d2):
    Cluster1x.append(data[i,0])
    Cluster1y.append(data[i,1])
    plt.plot(data[i,0],data[i,1],'bs')
  else:
    Cluster2x.append(data[i,0])
    Cluster2y.append(data[i,1])
    plt.plot(data[i,0],data[i,1],'gs')
print("datapoints in cluster1",len(Cluster1x))
print("datapoints in cluster2",len(Cluster2x))
```

```python
# auto clustering (2 cluster algorithm )
num=400
data = 0+(100-0)*np.random.random((num,2))
# consider datapoint0 as centroid 1
c1x = data[0,0];c1y = data[0,1]
# consider datapoint1 as centroid 2
c2x = data[1,0];c2y = data[1,1]

# step 2 find the distances of centroid and 1 and centroid 2 with respect to
# all the datapoints
for it in range(10):
  plt.figure(1,(10,2))
  plt.subplot(1,2,1)
  #plt.title('Raw data')
  plt.plot(data[:,0],data[:,1],'rs')
  Cluster1x=[];Cluster1y=[]
  Cluster2x=[];Cluster2y=[]
  for i in range(data.shape[0]):
    d1 = np.sqrt((c1x-data[i,0])**2 + (c1y-data[i,1])**2)
    d2 = np.sqrt((c2x-data[i,0])**2 + (c2y-data[i,1])**2)
    if(d1<d2):
      Cluster1x.append(data[i,0])
      Cluster1y.append(data[i,1])
      plt.plot(data[i,0],data[i,1],'bs')
    else:
      Cluster2x.append(data[i,0])
      Cluster2y.append(data[i,1])
      plt.plot(data[i,0],data[i,1],'gs')
  plt.plot(c1x,c1y,'kd') # to plot the centroid1
  plt.plot(c2x,c2y,'kd') # to display the centroid 2
  #print("datapoints in cluster1",len(Cluster1x))
  #print("datapoints in cluster2",len(Cluster2x))

  #step 3
  # now take the average of all the x coord and y coord of both
  # cluster 1 and cluster2
  C1xavg = sum(Cluster1x)/len(Cluster1x)
  C1yavg = sum(Cluster1y)/len(Cluster1y)

  C2xavg = sum(Cluster2x)/len(Cluster2x)
  C2yavg = sum(Cluster2y)/len(Cluster2y)

  # step4
  # update the centroids of cluster 1 and cluster 2 as per their averages
  c1x = C1xavg;c1y = C1yavg
  c2x = C2xavg;c2y = C2yavg
  plt.title(["Iteration number: "+str(it)])
  plt.pause(0.5) # half second delay
```
